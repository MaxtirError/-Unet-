# 数据可视化

先画出了训练中的Loss函数

可以任意读取result中predict的数据

通过Show函数得到图像在1-50部分的像素分布

通过pltshow展示图片，接下来直观地展示产生的噪声和去噪之后的结果。

```python
import numpy as np
import matplotlib.pyplot as plt
f = open('GlobalLoss.txt',"r")
Loss = [float(x.strip('\n')) for x in f.readlines()]
x_axis = range(len(Loss))
plt.plot(x_axis, Loss)
plt.show()
plt.close()
```

```python

def Show(src, color, label):
    x_aixs = range(1, 50)
    y_aixs = [0 for x in range(1, 50)]
    for x in src.ravel():
        x = max(min(int(x), 255), 0)
        if x != 0 and x < 50:
            y_aixs[x - 1] += 1
    plt.plot(x_aixs, y_aixs, c = color, label= label)
def pltshow(src):
    plt.imshow(src, cmap=plt.cm.gray, origin='lower')
    plt.show()
    plt.close()
plt.figure()
Origin = np.load('result\jdrw02010_drc_0_\Origin.npy')
noise = np.load(r'result\jdrw02010_drc_0_\noise_2.npy')
result = np.load(r'result\jdrw02010_drc_0_\result_2.npy')

Show(Origin, 'b', 'origin')
Show(noise, 'g', 'noise')
Show(result, 'r', 'result')
plt.legend()
plt.show()
plt.close()
pltshow(Origin)

origin_noise = abs(Origin - noise)*10
pltshow(origin_noise)


origin_result = abs(Origin - result)*10
pltshow(origin_result)

```
